# Setting up the Data

I will get ready to set up our data for our Random Forest Classification Model!


The **purpose** column as categorical

I need to transform it using dummy variables so sklearn will be able to understand them. 

I will create a list of 1 element containing the string 'purpose'.

```python
cat_feats = ['purpose']
```

I will create a fixed larger dataframe that has new feature columns with dummy variables.

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

```python
final_data.info()
```

## Train Test Split


```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop('not.fully.paid',axis=1)
y = final_data['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

## Training a Decision Tree Model


```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predictions and Evaluation of Decision Tree

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

## Training the Random Forest model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

## Predictions and Evaluation


```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
